```python
directory = "data/"

usa_events = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f):
        print(filename)
        name = filename.split("-")[2:]
        df = pd.read_csv(f)

        quarter = False
        if 'economic-growth-rate' in filename:
            quarter = True

        columns = list(df.columns)
        print(df.columns)

        annual_change_mean, annual_change_std = 0, 0
        if columns[2].strip() == 'Annual Change':
            annual_change_mean = df[columns[2]].mean()
            annual_change_std = df[columns[2]].std()

        df['date'] = pd.to_datetime(df['date'])
        df = df['1990-01-01' < df['date']] 
        df = df[df['date'] < '2024-01-01']
        print(df)

        # Iterate rows in df
        value = 0
        trail_years = []
        for index, row in df.iterrows():
            date = row['date']
            cur_value = row[columns[1]]
            print(date, cur_value)

            if value == 0:
                value = cur_value
            else:
                if cur_value > value:
                    if quarter:
                        print(date, "INCREASE QoQ", " ".join(name).upper()[:-4])
                        usa_events.append((date, "INCREASE QoQ", " ".join(name).upper()[:-4]))
                    else:
                        print(date, "INCREASE YoY", " ".join(name).upper()[:-4])
                        usa_events.append((date, "INCREASE YoY", " ".join(name).upper()[:-4]))
                    trail_years.append(1)
                else:
                    if quarter:
                        print(date, "DECREASE QoQ", " ".join(name).upper()[:-4])
                        usa_events.append((date, "DECREASE QoQ", " ".join(name).upper()[:-4]))
                    else:
                        print(date, "DECREASE YoY", " ".join(name).upper()[:-4])
                        usa_events.append((date, "DECREASE YoY", " ".join(name).upper()[:-4]))
                    trail_years.append(-1)
                if len(trail_years) > 5:
                    sumi = sum(trail_years[-5:])
                    if sumi == 5:
                        print(date, "INCREASE 5Y", " ".join(name).upper()[:-4])
                        usa_events.append((date, "INCREASE 5Y", " ".join(name).upper()[:-4]))
                    elif sumi == -5:
                        print(date, "DECREASE 5Y", " ".join(name).upper()[:-4])
                        usa_events.append((date, "DECREASE 5Y", " ".join(name).upper()[:-4]))
                if len(trail_years) > 3:
                    sumi = sum(trail_years[-3:])
                    if sumi == 3:
                        print(date, "INCREASE 3Y", " ".join(name).upper()[:-4])
                        usa_events.append((date, "INCREASE 3Y", " ".join(name).upper()[:-4]))
                    elif sumi == -3:
                        print(date, "DECREASE 3Y", " ".join(name).upper()[:-4])   
                        usa_events.append((date, "DECREASE 3Y", " ".join(name).upper()[:-4]))
                if len(trail_years) > 4 and quarter:
                    sumi = sum(trail_years[-4:])
                    if sumi == 4:
                        print(date, "INCREASE 4Q", " ".join(name).upper()[:-4])
                        usa_events.append((date, "INCREASE 4Q", " ".join(name).upper()[:-4]))
                    elif sumi == -4:
                        print(date, "DECREASE 4Q", " ".join(name).upper()[:-4])
                        usa_events.append((date, "DECREASE 4Q", " ".join(name).upper()[:-4]))
                if len(trail_years) > 8 and quarter:
                    sumi = sum(trail_years[-8:])
                    if sumi == 8:
                        print(date, "INCREASE 8Q", " ".join(name).upper()[:-4])
                        usa_events.append((date, "INCREASE 8Q", " ".join(name).upper()[:-4]))
                    elif sumi == -8:
                        print(date, "DECREASE 8Q", " ".join(name).upper()[:-4])
                        usa_events.append((date, "DECREASE 8Q", " ".join(name).upper()[:-4]))

            if annual_change_mean != 0:
                if row[' Annual Change'] > annual_change_mean + annual_change_std:
                    print(date, "LARGE INCREASE YoY", " ".join(name).upper()[:-4])
                    usa_events.append((date, "LARGE INCREASE YoY", " ".join(name).upper()[:-4]))
                elif row[' Annual Change'] < annual_change_mean - annual_change_std:
                    print(date, "LARGE DECREASE YoY", " ".join(name).upper()[:-4])
                    usa_events.append((date, "LARGE DECREASE YoY", " ".join(name).upper()[:-4]))

            value = cur_value
```

```python
class App:

    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        # Don't forget to close the driver connection when you are finished with it
        self.driver.close()

    def create_relation(self, name, relation, name2, relation_id, ts, ets):
        with self.driver.session(database="neo4j") as session:
            # Write transactions allow the driver to handle retries and transient errors
            result = session.execute_write(
                self._create_and_return_relation, name, relation, name2, relation_id, ts, ets)
            for row in result:
                print("Created node between: {p1}, {p2}".format(p1=row['p1'], p2=row['p2']))

    @staticmethod
    def _create_and_return_relation(tx, name, relation, name2, relation_id, ts, ets):
        # To learn more about the Cypher syntax, see https://neo4j.com/docs/cypher-manual/current/
        # The Reference Card is also a good resource for keywords https://neo4j.com/docs/cypher-refcard/current/
        
        query = (
            "MATCH (a:Country) WHERE a.name = $name "
            "RETURN a"
        )
        result = tx.run(query, name=name)

        if len(list(result)) == 0:
            print("Creating asset 1 ", name)
            query = (
            "CREATE (p1:Country { name: $name }) "
            )
            result = tx.run(query, name=name)

        query = (
            "MATCH (c:Economic_Indicator) WHERE c.name = $name2 "
            "RETURN c"
        )
        result = tx.run(query, name2=name2)
        
        if len(list(result)) == 0:
            query = (
                "CREATE (c1:Economic_Indicator { name: $name2 }) "
            )
            result = tx.run(query, name2=name2)

        query = (
            "MATCH (a:Country), (c:Economic_Indicator) WHERE a.name = $name AND c.name = $name2 "
            "CREATE (a)-[:"+relation+" { id: $relation_id, timestamp: $ts, expiry_timestamp: $ets }]->(c) "
            "RETURN a, c"
        )
        result = tx.run(query, name=name, name2=name2, relation_id=relation_id, ts=ts, ets=ets)
        try:
            return [{"p1": row["a"]["name"], "p2": row["c"]["name"]}
                    for row in result]
        # Capture any errors along with the query and data for traceability
        except ServiceUnavailable as exception:
            logging.error("{query} raised an error: \n {exception}".format(
                query=query, exception=exception))
            raise

    def find_person(self, person_name):
        with self.driver.session(database="neo4j") as session:
            result = session.execute_read(self._find_and_return_person, person_name)
            for row in result:
                print("Found person: {row}".format(row=row))

    @staticmethod
    def _find_and_return_person(tx, person_name):
        query = (
            "MATCH (p:Person) "
            "WHERE p.name = $person_name "
            "RETURN p.name AS name"
        )
        result = tx.run(query, person_name=person_name)
        return [row["name"] for row in result]
    
app = App(uri, user, password)
```
