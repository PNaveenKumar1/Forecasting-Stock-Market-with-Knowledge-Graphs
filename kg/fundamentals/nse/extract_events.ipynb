```python
ticker_to_df = {}


for ticker, fin_data in data.items():
    print(ticker)
    ii = 0
    for fin_df in fin_data:
        df = fin_df[0].T

        #print(filename, ii)
        ii += 1
        #if filename.split("-")[0] not in tickers_map:
        #    continue
        #print(len(df), df.columns)
        # Remove index 
        df = df.reset_index()

        date_map = {'Jan':1, "Feb":2, 'Mar':3, "Apr":4, "May":5, "Jun": 6, "Jul":7, "Aug":8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12}
        for idx, i in enumerate(df["index"]):
            if idx == 0:
                df["index"][idx] = "2008-01-01"
                continue            
            month, year = i.split("-")
            month = date_map[month]

            if "." in year:
                year = year.split(".")[0]
            year = 2000 + int(year)
            df["index"][idx] = str(year)+"-"+str(month)+"-30"

        date_col = "Year"
        if "Field" in df.columns:
            date_col = "Field"
        df["index"] = pd.to_datetime(df["index"])
        df = df[df["index"] >= pd.to_datetime("2008-01-01")]
        df = df.set_index("index")
        #Sort by date
        df = df.sort_index()

        
        #print(df)
        #print(df.columns)
        # If annual
        count1, count2 = 3, 5
        label1, label2, label = "3Y", "5Y", "YoY"
        expiry_months = 12
        if ii == 6:
            count1, count2 = 8, 16
            label1, label2, label = "8Q", "16Q", "QoQ"
            expiry_months = 3

        for columns in df.columns:
            history = []
            prev_value = None
            j = 0
            for rows in df.iterrows():
                if j == 0:
                    column_name = rows[1][columns]
                    #print(column_name)
                    j += 1
                    continue
                timestamp = rows[0]
                expiry = timestamp + pd.DateOffset(months=expiry_months)
                if rows[1][columns] == 'nan':
                    prev_value = None
                    history.append(0)
                    continue
                if prev_value == None:
                    prev_value = rows[1][columns]
                    history.append(0)
                else:
                    if prev_value < rows[1][columns]:
                        history.append(1)
                        #print(filename[:-4], "INCREASE QoQ", columns, timestamp, expiry)
                        if "annual" in filename:
                            financial_events.append((ticker, "INCREASE "+label, column_name, timestamp, expiry))
                    elif prev_value == rows[1][columns]:
                        history.append(0)
                    else:
                        history.append(-1)
                        #print(filename[:-4], "DECREASE QoQ", column_name, timestamp, expiry)
                        if "annual" in filename:
                            financial_events.append((ticker, "DECREASE "+label, column_name, timestamp, expiry))
                    if len(history) > count1:
                        if sum(history[-count1:]) == count1:
                            #print(ticker, "INCREASE "+label1, column_name, timestamp, expiry)
                            financial_events.append((ticker, "INCREASE "+label1, column_name, timestamp, expiry))
                        elif sum(history[-count1:]) == -count1:
                            #print(ticker, "DECREASE "+label1, column_name, timestamp, expiry)
                            financial_events.append((ticker, "DECREASE "+label1, column_name, timestamp, expiry))
                    if len(history) > count2:
                        if sum(history[-count2:]) == count2:
                            #print(ticker, "INCREASE "+label2, column_name, timestamp, expiry)
                            financial_events.append((ticker, "INCREASE "+label2, column_name, timestamp, expiry))
                        elif sum(history[-count2:]) == -count2:
                            #print(ticker, "DECREASE "+label2, column_name, timestamp, expiry)
                            financial_events.append((ticker, "DECREASE "+label2, column_name, timestamp, expiry))
                    if "quarterly" in filename:
                        if (prev_value - rows[1][columns]) / rows[1][columns] < -0.2:
                            #print(ticker, "LARGE INCREASE QoQ", column_name, timestamp, expiry)
                            financial_events.append((ticker, "LARGE INCREASE QoQ", column_name, timestamp, expiry))
                        elif (prev_value - rows[1][columns]) / rows[1][columns] > 0.2:
                            #print(ticker, "LARGE DECREASE QoQ", column_name, timestamp, expiry)
                            financial_events.append((ticker, "LARGE DECREASE QoQ", column_name, timestamp, expiry))
                    #    df.at[rows[0], columns] = 0
                    #print(columns, rows[1][columns])
                prev_value = rows[1][columns]
                
```

```python
class App:

    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        # Don't forget to close the driver connection when you are finished with it
        self.driver.close()

    def create_relation(self, name, relation, name2, relation_id, ts, ets):
        with self.driver.session(database="neo4j") as session:
            # Write transactions allow the driver to handle retries and transient errors
            result = session.execute_write(
                self._create_and_return_relation, name, relation, name2, relation_id, ts, ets)
            for row in result:
                print("Created node between: {p1}, {p2}".format(p1=row['p1'], p2=row['p2']))

    @staticmethod
    def _create_and_return_relation(tx, ticker, relation, name2, relation_id, ts, ets):
        # To learn more about the Cypher syntax, see https://neo4j.com/docs/cypher-manual/current/
        # The Reference Card is also a good resource for keywords https://neo4j.com/docs/cypher-refcard/current/

        query = (
            "MATCH (a:Asset), (c:Financial_Indicators) WHERE a.ticker = $ticker AND c.name = $name2 "
            "CREATE (a)-[:"+relation+" { id: $relation_id, timestamp: $ts, expiry_timestamp: $ets }]->(c) "
            "RETURN a, c"
        )
        result = tx.run(query, ticker=ticker, name2=name2, relation_id=relation_id, ts=ts, ets=ets)
        try:
            return [{"p1": row["a"]["ticker"], "p2": row["c"]["name"]}
                    for row in result]
        # Capture any errors along with the query and data for traceability
        except ServiceUnavailable as exception:
            logging.error("{query} raised an error: \n {exception}".format(
                query=query, exception=exception))
            raise

    def find_person(self, person_name):
        with self.driver.session(database="neo4j") as session:
            result = session.execute_read(self._find_and_return_person, person_name)
            for row in result:
                print("Found person: {row}".format(row=row))

    @staticmethod
    def _find_and_return_person(tx, person_name):
        query = (
            "MATCH (p:Person) "
            "WHERE p.name = $person_name "
            "RETURN p.name AS name"
        )
        result = tx.run(query, person_name=person_name)
        return [row["name"] for row in result]
    
app = App(uri, user, password)
```
